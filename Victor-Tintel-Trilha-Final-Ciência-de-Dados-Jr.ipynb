# - Próximo passo é  verificar os valores únicos de cada variável e isso é importante para analisarmos se vamos transformar uma variável em categórica ou não, porque uma variável pode estar como numérica e ser realmente um número mas podemos converter em uma variável categórica, para podermos fazer relação com outros dados.  

```python
valores_unicos_junho = []
for i in df_junho.columns[0:12].tolist():
    print(i, ":", len(df_junho[i].astype(str).value_counts()))
    valores_unicos_junho.append(len(df_junho[i].astype(str).value_counts()))
```

## - Então vamos criar uma lista vazia chamada de valores_unicos, em seguida vamos fazer um loop for , vamos varrer todas as colunas do nosso dataframe através do df_junho.columns e vamos colocar um slice de [0:12], isso porque o índice no python é exclusivo, então se vai até o 11, eu tenho que colocar um a mais edepois disso eu converto para uma lista através do .TOLIST().


## - Na linha de baixo eu estou exibindo a informação através do print, esse i que vai aparecer no print é o nome da coluna que o loop for está passando, portanto vai aparecer o nome de todas as colunas porque ele vai passar por todas elas.  O LEN é que eu vou passar por aquela coluna, astype str quer dizer que eu estou convertendo ela para string e o value_counts quer dizer que eu estou contando a quantidade de informações que eu tenho naquela coluna. Na última linha quer dizer que eu vou inserir a informações que eu peguei e adicionar através do append dentro da minha lista vazia VALORES_UNICOS.


```python
valores_unicos_julho = []
for i in df_julho.columns[0:12].tolist():
    print(i, ":", len(df_julho[i].astype(str).value_counts()))
    valores_unicos_julho.append(len(df_julho[i].astype(str).value_counts()))
```

# - Agora vamos analisar se existem outliers nos nossos dataframes, porém temos que lembrar que só vamos aplicar para as variáveis numéricas e como nós só temos uma coluna numérica, só será aplicado a ela. 

```python
# Carregamento das variaveis 

variaveis_numericas = []
for i in df_junho.columns[0:12].tolist():
    if df_junho.dtypes[i] == "int64" or df_junho.dtypes[i] == "float64":
        variaveis_numericas.append(i)    
```

```python
variaveis_numericas
```

- Aqui nesse caso colocamos escrito quais são as variáveis numéricas que temos, como só temos uma, só aparecu uma escrita.

```python
len(variaveis_numericas)
```

- Usamos o len para saber o tamanho dela, que nesse caso só temos uma variável numérica.

```python
plt.rcParams["figure.figsize"] = [18.00, 18.00]
plt.rcParams["figure.autolayout"] = True
f, axes = plt.subplots(2, 2) # 1 linha e 1 coluna

linha = 0
coluna = 0
for i in variaveis_numericas:
    sns.boxplot(data = df_junho, y=i, ax=axes[linha][coluna])
    

plt.show()
```

- De acordo com o nosso boxplot, temos dois valores outliers, porém não significa que realmente esse valor seja um outlier. Portanto temos que fazer uma análise mais detalhada para saber se é realmente uma discrepância nos dados ou não.
- Porém, fazendo uma análise de acordo com o groupby que fizemos anteriormente, nós temos divisões durante as horas semanais dedicadas de 1 hora até 28 horas e em diversas ocasiões durante esse tempo teve apenas uma opção de cada, por exemplo: com 10 horas feitas teve apenas uma pessoa, com 14 horas feita também teve e com isso com 28 horas feitas também só teve uma pessoa, portanto por mais que o boxplot conste tanto no mês de junho quanto no mês de julho alguns outliers, quando fazemos uma análise mais detalhada, todas as opções de 1 hora até 28 horas são importante serem analisadas, portanto podemos dizer que por mais que conste no boxplot, nós não vamos ter outliers nessa ocasião.

```python
plt.rcParams["figure.figsize"] = [18.00, 18.00]
plt.rcParams["figure.autolayout"] = True
f, axes = plt.subplots(2, 2) # 1 linha e 1 coluna

linha = 0
coluna = 0
for i in variaveis_numericas:
    sns.boxplot(data = df_julho, y=i, ax=axes[linha][coluna])
    

plt.show()
```

- A rimeira linha é o valor mínimo da coluna. A primeira linha da caixa é o primeiro quartil (25%), a segunda linha da caixa que é a linha que corta a caixa é o segundo quartil (50%) (que é a mediana), a linha que fecha a caixa é o terceiro quartil (75%) e a linha lá de cima é o valor máximo.
